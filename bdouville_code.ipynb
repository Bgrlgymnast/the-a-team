```python
# Function to split names correctly
def split_name(full_name):
    period_count = full_name.count('.')
    # Check if there are at least two periods
    if full_name.count('.') >= 2:
        # Split by the second period and keep only the part after the second period
        parts = full_name.split('.', 2)  # Max split into 3 parts
        first_name = parts[0] + '.' + parts[1]  # Combine first two parts
        last_name = parts[2]
    
    elif period_count == 1:
        # Case: Exactly one period, split by the first period
        parts = full_name.split('.', 1)
        first_name = parts[0]
        last_name = parts[1]
    else:   
        # If there are less than two periods, just split by space
        parts = full_name.split()
        first_name = parts[0]
        last_name = parts[-1]
    
    return pd.Series([first_name, last_name])
```

```python
# Apply the function to split the 'President' column
cleaned_pres_df[['first_name', 'last_name']] = cleaned_pres_df['President'].apply(split_name)
cleaned_pres_df

# Need to figure out a way to separate George H.W. Bush only one with 2 initials
```
