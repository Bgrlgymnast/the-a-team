```python
presidents_df= pd.read_csv("all_us_presidents.csv")
presidents_df.head()
```

```python
# get values for the years after 2013 for year column
add_years_after_inaguration = presidents_df["Years (after inauguration)"].iloc[-1]

# adding new rows that include Years (after inauguration), name of president and party
new_rows = {
    "Years (after inauguration)": [add_years_after_inaguration + i for i in range(1, 11)],
    "President": ['Barack Obama'] * 3 + ['Donald Trump'] * 4 + ['Joe Biden'] * 3,
    "Party": ['Democrat'] * 3 + ['Republican'] * 4 + ['Democrat'] * 3
}

#new_rows
```

```python
presidents_df.rename(columns={'Years (after inauguration)': 'Year'}, inplace=True)
presidents_df.head()
```

```python
updated_pres_df =pd.DataFrame(new_rows)
updated_pres_df
```

```python
updated_pres_df.rename(columns={'Years (after inauguration)': 'Year'}, inplace=True)
updated_pres_df
```

```python
updated_pres_df.rename(columns={'Years (after inauguration)': 'Year'}, inplace=True)
updated_pres_df
```

```python
cleaned_pres_df=pd.concat([presidents_df,updated_pres_df])
cleaned_pres_df.tail(15)
```

```python
cleaned_pres_df.loc[cleaned_pres_df["President"] == "George Washington", "Party"] = "Independent"
cleaned_pres_df.tail(25)
```

```python
# Function to split names correctly
def split_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = parts[-1]
    return pd.Series([first_name, last_name])
```

```python
# Apply the function to split the 'President' column
cleaned_pres_df[['First Name', 'Last Name']] = cleaned_pres_df['President'].apply(split_name)
cleaned_pres_df

# Need to figure out a way to separate George H.W. Bush only one with 2 initials
```

```python
# Drop the original 'President' column
cleaned_pres_df = cleaned_pres_df.drop(columns=['President'])
cleaned_pres_df
```

```python
cleaned_pres_df.to_csv("cleaned_pres_df.csv", index=False)
```
